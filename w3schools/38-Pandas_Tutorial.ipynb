### **معرفی پانداس**

پانداس (Pandas) یک کتابخانه محبوب در زبان برنامه‌نویسی پایتون برای کار با داده‌های ساختار یافته مانند جداول و فریم داده (DataFrames) است. این کتابخانه به توسعه‌دهندگان امکانات بسیاری برای مدیریت و تحلیل داده‌ها را فراهم می‌کند. در ادامه، به برخی از ویژگی‌ها و کاربردهای مهم پانداس می‌پردازم:

۱. فریم داده (DataFrame): پانداس یک مفهوم اصلی به نام "فریم داده" دارد که معادل یک جدول داده با ستون‌ها و ردیف‌ها است. این امکان را فراهم می‌کند تا بتوانید به راحتی با داده‌های خود کار کنید.

۲. خواندن و نویسی داده: پانداس از روش‌های مختلف برای خواندن داده از فرمت‌های مختلف مانند CSV، Excel، SQL و ... پشتیبانی می‌کند. همچنین امکان ذخیره داده به این فرمت‌ها نیز وجود دارد.

۳. انتخاب و فیلتر کردن داده: با استفاده از پانداس، می‌توانید به راحتی داده‌ها را بر اساس شرایط مشخص انتخاب و فیلتر کنید. این امکان به شما این اجازه را می‌دهد تا به داده‌های مورد نظرتان دسترسی پیدا کنید.

۴. ترتیب دادن و گروه‌بندی: می‌توانید داده‌ها را بر اساس مقادیر مشخص مرتب کنید یا آن‌ها را بر اساس ویژگی‌های خاص گروه‌بندی کنید.

۵. تغییر شکل داده (Reshaping): پانداس امکاناتی برای تغییر شکل داده‌ها، ادغام (merge) و ترکیب داده‌ها را فراهم می‌کند.

۶. محاسبات آماری: این کتابخانه به شما امکان محاسبات آماری مختلف نظیر میانگین، مجموع، حداکثر و حداقل را بر روی داده‌هایتان می‌دهد.

۷. رسم نمودارها: پانداس با استفاده از کتابخانه‌های دیگر مانند Matplotlib و Seaborn امکانات رسم نمودارهای مختلف را نیز دارد.

</section>

## **سری پانداس**

سری‌های پانداس مانند یک ستون در یک جدول است.
این یک آرایه یک بعدی برای نگهداری اطلاعات از هر نوعی می‌باشد.

</section>

```python
import pandas as pd

a = [1, 7, 2]

myVar = pd.Series(a)

print(myVar)
```

### **برچسب‌ها**

اگر چیز دیگری مشخص نشده باشد، مقادیر با شماره شاخص خود برچسب گذاری می شوند. مقدار اول دارای شاخص 0، مقدار دوم دارای شاخص 1 است.
از این برچسب می توان برای دسترسی به یک مقدار مشخص استفاده کرد.


</section>

```python
# Return the first value of the Series:

a = [1, 7, 2]

myVar = pd.Series(a)

print(myVar[0])
```

Create Labels

### **ساخت برچسب‌ها**

با آرگومان index، می توانید برچسب های خود را نام گذاری کنید.


</section>

```python
# Create your own labels:

import pandas as pd

a = [1, 7, 2]

myVar = pd.Series(a, index=["x", "y", "z"])

print(myVar)
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

همچنین می‌توانید از یک شی کلید/مقدار، مانند دیکشنری (dictionary)، هنگام ایجاد یک سری استفاده کنید.

</section>

```python
# Create a simple Pandas Series from a dictionary:

calories = {"day1": 420, "day2": 380, "day3": 390}

myVar = pd.Series(calories)

print(myVar)
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

⚠️ توجه: کلیدهای فرهنگ لغت به برچسب تبدیل می شوند.

برای انتخاب فقط برخی از موارد موجود در فرهنگ لغت، از آرگومان شاخص استفاده کنید و فقط مواردی را که می خواهید در سری گنجانده شود، مشخص کنید.

</section>

```python
# Create a Series using only data from "day1" and "day2":

calories = {"day1": 420, "day2": 380, "day3": 390}

myVar = pd.Series(calories, index=["day1", "day2"])

print(myVar)
```

## **پاکسازی سلول‌های خالی**

سلول‌های خالی به طور بالقوه می‌توانند هنگام تجزیه و تحلیل داده‌ها، نتیجه اشتباهی به شما بدهند.


### **حذف ردیف‌های خالی**

یکی از راه های مقابله با سلول های خالی حذف ردیف هایی است که حاوی سلول های خالی هستند.
این معمولاً مشکلی ندارد، زیرا مجموعه داده‌ها می‌توانند بسیار بزرگ باشند و حذف چند ردیف تأثیر زیادی در نتیجه نخواهد داشت.

</section>

```python
import pandas as pd
import os

folder_name = 'src'
file_name = 'data.csv'

file_path = os.path.join(folder_name, file_name)

df = pd.read_csv(file_path)

new_df = df.dropna()

print(new_df.to_string())
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

⚠️ توجه: به طور پیش‌فرض، متد dropna() یک DataFrame جدید برمی‌گرداند و نسخه اصلی را تغییر نخواهد داد.

اگر می خواهید DataFrame اصلی را تغییر دهید، از آرگومان inplace = True استفاده کنید:


</section>

```python
df.dropna(inplace=True)

print(df.to_string())
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

⚠️ توجه: اکنون dropna (inplace = True) DataFrame جدیدی را برنمی‌گرداند، اما تمام ردیف‌های حاوی مقادیر NULL را از DataFrame اصلی حذف می‌کند.

</section>

### **فقط برای ستون های مشخص شده جایگزین کنید**

مثال بالا جایگزین تمام سلول های خالی در کل Data Frame می شود.
فقط برای جایگزینی مقادیر خالی برای یک ستون، نام ستون را برای DataFrame مشخص کنید:

</section>

```python
df["Calories"].fillna(130, inplace=True)
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

یک روش رایج برای جایگزینی سلول های خالی، محاسبه میانگین، میانه یا مقدار حالت ستون است.
پانداس از متدهای mean() median() و mode() برای محاسبه مقادیر مربوطه برای یک ستون مشخص استفاده می کنند:

</section>

```python
import pandas as pd
import os

folder_name = 'src'
file_name = 'data.csv'

file_path = os.path.join(folder_name, file_name)

df = pd.read_csv(file_path)

x = df["Calories"].mean()

df["Calories"].fillna(x, inplace=True)

# print(df.head())
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

⚠️ میانگین = مقدار متوسط ​​(مجموع همه مقادیر تقسیم بر تعداد مقادیر).

</section>

```python
import pandas as pd
import os

folder_name = 'src'
file_name = 'data.csv'

file_path = os.path.join(folder_name, file_name)

df = pd.read_csv(file_path)

x = df["Calories"].median()

df["Calories"].fillna(x, inplace=True)
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

⚠️ میانه = مقدار در وسط، بعد از اینکه همه مقادیر را به صورت صعودی مرتب کردید.

</section>

```python
import pandas as pd
import os

folder_name = 'src'
file_name = 'data.csv'

file_path = os.path.join(folder_name, file_name)

df = pd.read_csv(file_path)

x = df["Calories"].mode()[0]

df["Calories"].fillna(x, inplace=True)
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

⚠️ Mode = مقداری که بیشتر ظاهر می شود.

</section>

## **پاکسازی داده‌های با فرمت اشتباه**

سلول هایی با داده هایی با فرمت اشتباه می توانند تجزیه و تحلیل داده ها را دشوار یا حتی غیرممکن کنند.

برای رفع آن، دو گزینه دارید: ردیف‌ها را حذف کنید یا تمام سلول‌های ستون‌ها را به یک قالب تبدیل کنید.

در Data Frame ما دو سلول با فرمت اشتباه داریم. سطر 22 و 26 را بررسی کنید، ستون 'Date' باید رشته ای باشد که تاریخ را نشان می دهد:

## **اصلاح داده‌های اشتباه**

Replacing Values

### **جایگزینی مقادیر**

یکی از راه‌های اصلاح مقادیر اشتباه، جایگزینی آنها با چیز دیگری است.
در مثال ما، به احتمال زیاد اشتباه تایپی است، و مقدار باید به جای "450" ​​"45" باشد، و ما فقط می توانیم "45" را در ردیف 7 وارد کنیم:

</section>

```python
df.loc[7, 'Duration'] = 45
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

برای مجموعه‌های داده‌های کوچک، ممکن است بتوانید داده‌های اشتباه را یکی یکی جایگزین کنید، اما نه برای مجموعه‌های داده بزرگ.
برای جایگزینی داده های اشتباه برای مجموعه داده های بزرگتر، می توانید قوانینی ایجاد کنید، به عنوان مثال. برخی از مرزها را برای مقادیر قانونی تعیین کنید و هر مقداری را که خارج از مرزها است جایگزین کنید.

</section>

```python
# If the value is higher than 120, set it to 120:

for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

### **حذف ردیف‌ها**

روش دیگر مدیریت داده های اشتباه حذف ردیف هایی است که حاوی داده های اشتباه هستند.
به این ترتیب شما مجبور نیستید بفهمید که با چه چیزی جایگزین کنید و احتمال زیادی وجود دارد که برای انجام تجزیه و تحلیل خود به آنها نیاز نداشته باشید.

</section>

```python
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.drop(x, inplace=True)
```

⚠️ نکته: متد ()corr ستون های "غیر عددی" را نادیده می گیرد.

نتیجه متد ()corr یک جدول با تعداد زیادی اعداد است که نشان می دهد چقدر رابطه بین دو ستون خوب است.

-0.9 به اندازه 0.9 رابطه خوبی خواهد بود، اما اگر یک مقدار را افزایش دهید، مقدار دیگر احتمالاً پایین می آید.
0.2 به معنای رابطه خوبی نیست، به این معنی که اگر یک مقدار بالا رفت به این معنی نیست که مقدار دیگر افزایش می یابد.

⚠️ همبستگی خوب چیست؟ بستگی به کاربرد دارد، اما فکر می‌کنم می‌توان گفت که باید حداقل 0.6 (یا -0.6) داشته باشید تا آن را یک همبستگی خوب بدانیم.
