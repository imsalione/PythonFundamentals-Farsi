```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder_name = 'src'
file_name = 'magic04.data'
file_path = os.path.join(folder_name, file_name)

cols = cols = ['fLenght', 'fWidth', 'fSize', 'fConc', 'fConc1',
               'fAsym', 'fM3Long', 'fM3Trans', 'fAlpha', 'fDist', 'class']
df = pd.read_csv(file_path, names=cols)
df.head()
```

```python
df['class'] = (df['class'] == 'g').astype(int)
df.head()
```

```python
for label in cols[:-1]:
    plt.hist(df[df['class'] == 1][label], color='blue',
             label='gamma', alpha=0.7, density=True)
    plt.hist(df[df['class'] == 0][label], color='red',
             label='hadron', alpha=0.7, density=True)
    plt.title(label)
    plt.ylabel('Probability')
    plt.xlabel(label)
    plt.legend()
    plt.show()
```
